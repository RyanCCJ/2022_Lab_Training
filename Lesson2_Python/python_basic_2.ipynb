```python
# test
#自己用open創.txt，一行一條序列(ATCG等等)，輸出互補後的序列成.txt並也一行一條序列(ATCG -> TAGC)
with open('test_0709.txt', 'w') as f:
    f.write('TTAAGGC\n')
    f.write('ATGCGCTA\n')

with open('test_0709.txt', 'r') as f:
    data = f.readlines()
    
print(data)
r_list = []
for i in range(len(data)):
    r_seq = data[i].replace('\n', '').replace('A', 'T').replace('T', 'A').replace('C', 'G').replace('G', 'C')
    r_seq = r_seq.upper()
    r_list.append(r_seq)
    
with open('output_0709.txt', 'w') as f:
    for seq in r_list:
        f.write(seq)
        f.write('\n')
```

```python
# pass/continue/break
# break：強制跳出 ❮整個❯ 迴圈
# continue：強制跳出 ❮本次❯ 迴圈，繼續進入下一圈
# pass：不做任何事情，所有的程式都將繼續

for i in range(0,5):
    if i == 3:
        pass
        #print(i)
    print(i)
```

```python
for i in range(0,5):
    if i == 3:
        continue
    print(i)
```

```python

for i in range(0,5):
    if i == 3:
        break
    print(i)
```
